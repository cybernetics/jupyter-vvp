# Some button examples

Showing indicators that can update themselves using the `jupyter_interval_widget` module:

```python
import ipywidgets as widgets

from jupyter_interval_widget import Interval

from IPython.display import display, clear_output
import random
import string

def random_string(n=4):
    return ''.join(random.choices(string.digits + "abcdef", k=n))

output = widgets.Output()

def get_deployment_status():
    if random.randint(0,1):
        return "success"
    return "danger"

def update_output(self):
    print("Updating output")
    with output:
        clear_output(wait=True)
        status = get_deployment_status()
        print("Status: {}.".format(status))
    status_button.button_style=status
    status_button.description = status
           
status_button = widgets.Button(
                description="Will soon refresh",
                disabled=True,
                button_style="info", # 'success', 'info', 'warning', 'danger' or ''
                tooltip='Click to refresh deployment status',
                icon='refresh' # (FontAwesome names without the `fa-` prefix)
         )


def stop_listening(b):
    print("Stopping updates")
    interval.close()
    stop_button.disabled=True
           
stop_button = widgets.Button(
                description="Click to stop",
                disabled=False,
                button_style="danger", # 'success', 'info', 'warning', 'danger' or ''
                tooltip='Click to stop updating',
                icon='stop' # (FontAwesome names without the `fa-` prefix)
         )
stop_button.on_click(stop_listening)

interval = Interval(value=2000)
interval.on_tick(update_output)

display(widgets.HBox(children=(output, interval, status_button, stop_button)))
```

Below is a clickable button:

```python
import ipywidgets as widgets

from IPython.display import display, clear_output
import random
import string
import time

def random_string(n=4):
    return ''.join(random.choices(string.digits + "abcdef", k=n))

output = widgets.Output()

def get_deployment_status():
    time.sleep(1)
    if random.randint(0,1):
        return "success"
    return "danger"


def update_button(b):
    with output:
        button.disabled=True
        clear_output(wait=True)
        print("Getting status...")
        status = get_deployment_status()
        clear_output(wait=True)
        print("Status: {}.".format(status))
        button.button_style=status
        button.disabled=False

button = widgets.Button(
                description="Refresh",
                disabled=False,
                button_style="info", # 'success', 'info', 'warning', 'danger' or ''
                tooltip='Click to refresh deployment status',
                icon='refresh' # (FontAwesome names without the `fa-` prefix)
            )            
button.on_click(update_button)
with output:
    print("Status: transitioning.")
display(widgets.HBox(children=(button,output)))
```
